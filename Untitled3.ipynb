```
#m√©todo de Euler Progresivo
import numpy as np
from matplotlib import pyplot
#Se definen las condiciones iniciales de posicion, velocidad y tiempo:
t_0 = 0
x_0 = 10.0
y_0 = 0.0
vx_0 = 0.0
vy_0 = 0.3
#Se define el tiempo de termino 
t_f = 180.00
#Se define una cantidad fija de intervalos
n = 3000.0
N = 3000
#Definimos los delta t para determinar velocidades
dt=(t_f-t_0)/(n-1)
#creamos un vector de puros 0
x = np.zeros([N])
y = np.zeros([N])
vx = np.zeros([N])
vy = np.zeros([N])
y[0] = y_0
x[0] = x_0
vx[0] = vx_0
vy[0] = vy_0
for i in range(1,N):
    vx[i] = dt*(-x[i-1]/(((x[i-1])**2 + y[i-1]**2)**(3/2))) + vx[i-1] #Aca esta la iteracion para las velocidades
    vy[i] = dt*(-y[i-1]/(((x[i-1])**2 + y[i-1]**2)**(3/2))) + vy[i-1]
    x[i] = x[i-1] +dt*vx[i-1] #Esta es la iteracion para las posicions
    y[i] = y[i-1] + dt*vy[i-1]
pyplot.plot(x,y)
pyplot.xlabel('x')
pyplot.ylabel('y')
pyplot.axhline(0, color="black")
pyplot.axvline(0, color="black")
# Limitar los valores de los ejes.


pyplot.savefig("output.png")

pyplot.show()
    
   
```
